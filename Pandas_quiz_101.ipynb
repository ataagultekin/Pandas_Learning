```python
import numpy as np
import seaborn as sns
import pandas as pd
#pd.set_option('display.max_rows', None)
pd.set_option("display.max_columns",None)
```

```python
# Task 1: Read the Titanic dataset from the Seaborn library.

# Görev 1: Seaborn kütüphanesi içerisinden Titanic veri setini tanımlayınız.
df = sns.load_dataset("titanic")
df
```

```python
# Task 2: Find the number of male and female passengers in the Titanic dataset. 
# Görev 2: Yukarıda tanımlanan Titanic veri setindeki kadın ve erkek yolcuların sayısını bulunuz.

df[["sex"]].value_counts()
```

```python
# Task 3: Find the number of unique values for each column.
# Görev 3: Her bir sutuna ait unique değerlerin sayısını bulunuz.

df.nunique()
```

```python
# Task 4: Find the unique values of the variable pclass.
# Görev 4: pclass değişkeninin unique değerlerini bulunuz.
#ascending sort True küçük büyük false büyük küçük

df=df.sort_values(by=['pclass'], ascending=True)
df["pclass"].unique()
```

```python
# Task 5: Find the number of unique values of pclass and parch variables. 
# Görev 5: pclass ve parch değişkenlerinin unique değerlerinin sayısını bulunuz.

df[["pclass","parch"]].nunique()
```

```python
# Task 6: Check the type of the embarked variable. Change its type to category. Check the type, again.
# Görev 6: embarked değişkeninin tipini kontrol ediniz. Tipini category olarak değiştiriniz. Tekrar tipini kontrol ediniz.

df["embarked"] = df["embarked"].astype("category")
df["embarked"].dtype
```

```python
# Task 7: Show all the information whose embarked value is C.
# Görev 7: embarked değeri C olanların tüm bilgelerini gösteriniz.

df[df["embarked"] =="C"]
```

```python
# Task 8: Show all the information whose embarked value is not S.

# Görev 8: embarked değeri S olmayanların tüm bilgelerini gösteriniz.

df[df["embarked"] !="S"]
```

```python
# Task 9: Show all information for female passengers younger than 30 years old.
# Görev 9: Yaşı 30 dan küçük ve kadın olan yolcuların tüm bilgilerini gösteriniz.

df[(df["sex"] !="male")&(df["age"]<30)]
```

```python
# Task 10: Show the information of passengers whose Fare is over 500 or 70 years old.
# Görev 10: Fare'i 500'den büyük veya yaşı 70 den büyük yolcuların bilgilerini gösteriniz.

df[(df["fare"] >500)|(df["age"]>70)]
```

```python
# Task 11: Find the null values/missing values in each variable. 
# Görev 11: Her bir değişkendeki boş değerlerin toplamını bulunuz.

df.isnull().sum()
```

```python
# Task 12: Drop the who variable from the dataframe.
# Görev 12: who değişkenini dataframe'den düşürün.

df.drop(["who"], axis = 1)
```

```python
# Task 13: Fill the missing values in the deck variable with the most repeated value (mode) of the deck variable. 
# Görev 13: deck değişkenindeki boş değerleri deck değişkenin en çok tekrar eden değeri (mode) ile doldurunuz.

x = df[["deck"]].mode()
x
# değeri değiştirmek için tek köşeli parantez kullanıyoruz(caner abiye sor )7
# axis = 0 demek row row değiştir, axis = 1 colum değişitir
df["deck"].fillna("C",axis = 0, inplace = True)
#Boş değer check etmek için

df[["deck"]].isnull().sum()
```

```python
# Task 14: Fill the missing values in the age variable with the median of the age variable.

# Görev 14: age değikenindeki boş değerleri age değişkenin medyanı ile doldurun.

df["age"].fillna(df["age"].median(),axis=0, inplace=True)

df["age"].isnull().sum()
```

```python
# Task 15: Find the sum, count, mean values of the survived variable in the breakdown of the pclass and gender variables. 
# Görev 15: survived değişkeninin Pclass ve Cinsiyet değişkenleri kırılımınında sum, count, mean değerlerini bulunuz.

df.groupby(["pclass","sex"]).agg({"survived":["sum","count","mean"]}).reset_index()
```

```python
# Task 16: Write a function that returns 1 for whose age is under 30, 0 for whose age is greater than or equal to 30.
# Create a variable named age_flag in the titanic data set using the function you wrote.
# (use apply and lambda functions)


# Görev 16: 30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 vericek bir fonksiyon yazınız.
# Yazdığınız fonksiyonu kullanarak titanik veri setinde age_flag adında bir değişken oluşturunuz oluşturunuz. 
# (apply ve lambda yapılarını kullanınız)


df["age_flag"]=df["age"].apply(lambda x:1 if x<30 else 0)
df.head()
```

```python
# Task 17: Read the Tips dataset from the Seaborn library.
# Görev 17: Seaborn kütüphanesi içerisinden Tips veri setini tanımlayınız.

tips = sns.load_dataset("tips")
tips
```

```python
# Task 18: Find the sum, min, max and average of total_bill values 
# with respect to the categories (Dinner, Lunch) of Time variable.

# Görev 18: Time değişkeninin kategorilerine (Dinner, Lunch) göre total_bill  
# değerlerinin toplamını, min, max ve ortalamasını bulunuz.
tips.groupby(["time"]).agg({"total_bill":["sum","min","max","mean"]}).reset_index()
```

```python
# Task 19: Find the sum, min, max and average of total_bill values by days and time.
# Görev 19: Günlere ve time göre total_bill değerlerinin toplamını, min, max ve ortalamasını bulunuz.
tips.groupby(["day","time"]).agg({"total_bill":["sum","min","max","mean"]}).reset_index()
```

```python
# Task 20: Find the sum, min, max and average values of the total_bill and tip variables according to day
# for Lunch time and female customers.
# Görev 20:Lunch zamanına ve kadın müşterilere ait total_bill ve tip  
# değerlerinin day'e göre toplamını, min, max ve ortalamasını bulunuz.

tips[(tips["time"]== "Lunch") & (tips["sex"]== "Female")].groupby(["day"]).agg({"total_bill":["sum","min","max","mean"],"tip":["sum","min","max","mean"]})
```

```python
# Task 21: What is the average of orders with size less than 3 and total_bill greater than 10?
# Görev 21: size'i 3'ten küçük, total_bill'i 10'dan büyük olan siparişlerin ortalaması nedir?

tips[(tips["size"]<3)&(tips["total_bill"]>10)].agg(["mean"])
```

```python
# Task 22: Create a new variable called total_bill_tip_sum,
# which gives the sum of the total bill and tip paid by each customer.


# Görev 22: total_bill_tip_sum adında yeni bir değişken oluşturun. 
# Her bir müşterinin ödediği totalbill ve tip in toplamını versin.

tips["total_bill_tip_sum"]=tips["total_bill"]+tips["tip"]
tips.head()
```

```python
# Task 23: Find the mean of the total_bill variable for men and women separately.
# 0 is given to those whose averages less than you have found, and 1 is given to those whose averages greater than or equal to 
# you have found.
# Create a new total_bill_flag variable.
# Attention !! For females, the average for females will be taken into account, and for male, the average for males will 
# be taken into account.
# Start by writing a function that takes a gender and total_bill as a parameter (will include if-else conditions).

# Görev 23: Total_bill değişkeninin kadın ve erkek için ayrı ayrı ortalamasını bulun.
# Bulduğunuz ortalamaların altında olanlara 0, üstünde ve eşit olanlara bir verildiği 
# yeni bir total_bill_flag değişkeni oluşturun.
# Dikkat !! Female olanlar için kadınlar için bulunan ortalama dikkate alıncak, male için ise erkekler için bulunan ortalama.
# parametre olarak cinsiyet ve total_bill alan bir fonksiyon yazarak başlayın. (If-else koşulları içerecek)

tips["total_bill_flag"] = tips["total_bill"].apply(lambda x: 0  if x<tips[(tips["sex"]=="Male")|(tips["sex"]=="Female")]["total_bill"].mean()else 1)
```

```python
# Task 24: Using the total_bill_flag variable, find the number of observations whose average is smaller or greater 
# than what you have found.

# Görev 24: total_bill_flag değişkenini kullanarak cinsiyetlere göre ortalamanın 
# altında ve üstünde olanların sayısını gözlemleyin.
tips.head()
tips.groupby(["sex","total_bill_flag"]).agg({"total_bill_flag":["count"]})
```

```python
# Task 25: Sort from largest to smallest according to the total_bill_tip_sum variable 
# and assign the first 30 people to a new dataframe.


# Görev 25: total_bill_tip_sum değişkenine göre büyükten küçüğe sıralayınız ve ilk 30 kişiyi yeni bir dataframe'e atayınız.
tips["total_bill_tip_sum"]=tips["total_bill"]+tips["tip"]
tips.sort_values(by=['total_bill_tip_sum'], ascending=True,inplace=True)
tip_s = tips.head(30)
tip_s.count()
```
